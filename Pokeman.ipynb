## Filtering Data

```python
new_df=df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] =='Poison') & ((df['HP'] >= 70) | (df['HP']<=30))]

new_df = new_df.reset_index(drop=True)
new_df
```

```python
## We can filter the Name by not allowing it to include "Mega"
df.loc[df['Name'].str.contains('Mega')]

## If we do not want the name containing "Mega", we do the following 
df.loc[~df['Name'].str.contains('Mega')]

## Suppose we want to get if Type 1 was either "Grass" or "Fire"
import re ##Import regular expressions
df.loc[df['Type 1'].str.contains('fire|grass',flags=re.I,regex=True)]

## Get all pokemon names started with a "Pi"
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)]
```

### Aggregate Statistics (GroupBy)

```python
df = pd.read_csv("Modified.csv")

## Check the avaerge for Type 1
df.groupby(['Type 1']).mean()

## Sort the Type 1
df.groupby(['Type 1']).mean().sort_values('HP',ascending=False)

## Check the sum for Type 1
df.groupby(['Type 1']).sum()

## Check the count for Type 1
df.groupby(['Type 1']).count()
```

### Working with Large amounts of data

```python
## When working with a larger dataset
## We load it this way

## The chunk means that 5 rows are being passed at a time
#for df in pd.read_csv('Modified.csv',chunksize=5):
#    print("CHUNK DF")
#    print(df)


new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv("Modified.csv",chunksize=5):
    results = df.groupby(['Type 1']).count()
    new_df = pd.concat([new_df,results])

new_df

```
